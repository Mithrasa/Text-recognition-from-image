```python
import cv2
import pytesseract
from google.colab.patches import cv2_imshow

def execute_part3(image_path):
                                                                                # Load the input image
    image = cv2.imread(image_path)

                                                                                # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    preprocessed = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]

                                                                                # Perform text detection with PyTesseract
    data = pytesseract.image_to_data(preprocessed, lang='eng', config='--psm 6', output_type=pytesseract.Output.DICT)

                                                                                # Iterate over the detected words and draw bounding boxes
    for i in range(len(data['text'])):
                                                                                # Only consider words with a confidence score above a threshold 
        if int(data['conf'][i]) > 60:
            x, y, w, h = data['left'][i], data['top'][i], data['width'][i], data['height'][i]
            cv2.rectangle(image, (x, y), (x + w, y + h), (0, 255, 0), 2)
            cv2.putText(image, data['text'][i], (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

                                                                                # Display the output image with bounding boxes and annotations
    cv2_imshow(image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

image_path1 = "/content/drive/MyDrive/input3.jpeg"
image_path2 = "/content/drive/MyDrive/input6.png"
image_path3 = "/content/drive/MyDrive/input8.jpeg"
image_path4 = "/content/drive/MyDrive/input7.jpeg"

im_list = [image_path1,image_path2,image_path3,image_path4]
for i in im_list:
  execute_part3(i)
```
